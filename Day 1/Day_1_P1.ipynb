```python
from transformers import pipeline

#1) create a sentiment analysis pipeline using a ready-made model
classifier = pipeline("sentiment-analysis",
                      model="distilbert-base-uncased-finetuned-sst-2-english")

#2) inference
text = input("Enter a text to Analyse:")
result = classifier(text) #returns a list of dicts like [{'label' : 'positive'}]
print(result)
```

```python
from transformers import pipeline

qa = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

context = (
    "The Apollo 11 mission landed on the moon."
    "Neil Armstrong walked on the moon"
    )
question = "Who walked on the moon?"
result = qa(question=question, context=context)
print(result)
```

```python
#Load model directly
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM


checkpoint="codeboosterstech/EN-TA"
tokenizer=AutoTokenizer.from_pretrained(checkpoint)
model=AutoModelForSeq2SeqLM.from_pretrained(checkpoint)


def language_translator(text):
  tokenized=tokenizer([text],return_tensors='pt')
  out=model.generate(**tokenized,max_length=128)
  return tokenizer.decode(out[0],skip_special_tokens=True)


text=input("Enter a text to translate:")
print(language_translator(text))

```

```python
import os
import gradio as gr
from huggingface_hub import InferenceClient

# -------------------------------
# Model configuration
# -------------------------------
MODEL_ID = "mistralai/Mistral-7B-Instruct-v0.2"

client = InferenceClient(
    model=MODEL_ID,
    api_key=os.environ["HF_TOKEN"]  # Set HF_TOKEN in environment
)

# -------------------------------
# Helper function to extract text
# -------------------------------
def extract_text(resp):
    try:
        return resp.choices[0].message.content
    except Exception:
        try:
            return resp["choices"][0]["message"]["content"]
        except:
            return str(resp)

# -------------------------------
# Core LLM function
# -------------------------------
def leetcode_solver(user_question):
    messages = [
        {
            "role": "system",
            "content": "You are a helpful Leetcode problem solver. Explain answers simply with Python examples."
        },
        {
            "role": "user",
            "content": user_question
        }
    ]

    response = client.chat_completion(
        messages=messages,
        max_tokens=500
    )

    return extract_text(response)

# -------------------------------
# Gradio UI
# -------------------------------
with gr.Blocks(title="LeetCode AI Tutor") as demo:
    gr.Markdown("## LeetCode AI Tutor (Mistral)")
    gr.Markdown("Ask any LeetCode problem and get a **simple Python explanation**.")

    input_box = gr.Textbox(
        label="Enter LeetCode Problem / Question",
        placeholder="Explain Two Sum problem with Python code",
        lines=3
    )

    output_box = gr.Textbox(
        label="AI Explanation",
        lines=15
    )

    submit_btn = gr.Button("Solve")

    submit_btn.click(
        fn=leetcode_solver,
        inputs=input_box,
        outputs=output_box
    )

# -------------------------------
# Launch app
# -------------------------------
demo.launch()
```
