```python
from langchain_openai import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain, SequentialChain
from twilio.rest import Client

llm = OpenAI(temperature=0.3)


# -----------------------------
# Single LLM (Mandatory)
# -----------------------------
llm = OpenAI(temperature=0.3)

# -----------------------------
# Chain 1: Input Understanding
# -----------------------------
prompt1 = PromptTemplate(
    input_variables=["raw_input"],
    template="""
    Understand the following instruction and extract its purpose clearly:
    Instruction: {raw_input}
    Cleaned Intent:
    """
)

chain1 = LLMChain(llm=llm, prompt=prompt1, output_key="intent")

# -----------------------------
# Chain 2: Content Generation
# -----------------------------
prompt2 = PromptTemplate(
    input_variables=["intent"],
    template="""
    Generate a clear and professional academic advisory message
    based on the following intent:
    {intent}
    Message:
    """
)

chain2 = LLMChain(llm=llm, prompt=prompt2, output_key="message")

# -----------------------------
# Chain 3: WhatsApp Optimization
# -----------------------------
prompt3 = PromptTemplate(
    input_variables=["message"],
    template="""
    Optimize the following message for WhatsApp delivery.
    Keep it short, polite, and student-friendly:
    {message}
    Final WhatsApp Text:
    """
)

chain3 = LLMChain(llm=llm, prompt=prompt3, output_key="final_text")

# -----------------------------
# Sequential Agent
# -----------------------------
agent = SequentialChain(
    chains=[chain1, chain2, chain3],
    input_variables=["raw_input"],
    output_variables=["final_text"]
)

# -----------------------------
# Twilio WhatsApp Sender
# -----------------------------
def send_whatsapp(message, to_number):
    client = Client()
    client.messages.create(
        body=message,
        from_="whatsapp:+14155238886",  # Twilio Sandbox Number
        to=f"whatsapp:{to_number}"
    )

# -----------------------------
# Run Agent
# -----------------------------
user_instruction = "Remind final year students about internship report submission tomorrow"
result = agent.invoke({"raw_input": user_instruction})

send_whatsapp(result["final_text"], "+91XXXXXXXXXX")

print("WhatsApp message sent successfully!")
```

```python
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain, SequentialChain
from twilio.rest import Client

# ----------------------------------
# SINGLE LLM (MANDATORY REQUIREMENT)
# ----------------------------------
llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0.3
)

# ----------------------------------
# CHAIN 1: INPUT UNDERSTANDING
# ----------------------------------
prompt_1 = PromptTemplate(
    input_variables=["raw_input"],
    template="""
    Understand the instruction and clearly extract its intent.

    Instruction: {raw_input}
    Intent:
    """
)

chain_1 = LLMChain(
    llm=llm,
    prompt=prompt_1,
    output_key="intent"
)

# ----------------------------------
# CHAIN 2: MESSAGE GENERATION
# ----------------------------------
prompt_2 = PromptTemplate(
    input_variables=["intent"],
    template="""
    Create a professional academic advisory message
    based on the intent below:

    Intent: {intent}
    Message:
    """
)

chain_2 = LLMChain(
    llm=llm,
    prompt=prompt_2,
    output_key="message"
)

# ----------------------------------
# CHAIN 3: WHATSAPP OPTIMIZATION
# ----------------------------------
prompt_3 = PromptTemplate(
    input_variables=["message"],
    template="""
    Convert the message into a short, polite,
    WhatsApp-friendly format for students.

    Message: {message}
    Final WhatsApp Text:
    """
)

chain_3 = LLMChain(
    llm=llm,
    prompt=prompt_3,
    output_key="final_text"
)

# ----------------------------------
# SEQUENTIAL AGENT
# ----------------------------------
agent = SequentialChain(
    chains=[chain_1, chain_2, chain_3],
    input_variables=["raw_input"],
    output_variables=["final_text"]
)

# ----------------------------------
# TWILIO WHATSAPP FUNCTION
# ----------------------------------
def send_whatsapp(message, to_number):
    client = Client()
    client.messages.create(
        body=message,
        from_="whatsapp:+14155238886",   # Twilio Sandbox Number
        to=f"whatsapp:{to_number}"
    )

# ----------------------------------
# RUN THE AGENT
# ----------------------------------
instruction = "Remind final year students to submit their internship report tomorrow"
result = agent.invoke({"raw_input": instruction})

send_whatsapp(
    message=result["final_text"],
    to_number="+91XXXXXXXXXX"   # Replace with student WhatsApp number
)

print("✅ WhatsApp message sent successfully!")
```
