```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

# -------------------------------
# Initialize Groq LLM
# -------------------------------
llm = ChatGroq(
    groq_api_key="gsk_xxxxxx",
    model="llama-3.1-8b-instant",
    temperature=0
)

parser = StrOutputParser()

# -------------------------------
# STEP 1: Generate Multiple Thoughts
# -------------------------------
thought_prompt = PromptTemplate(
    input_variables=["problem"],
    template="""
You are an expert problem solver.

Generate THREE different solution strategies for the problem below.
Do not solve yet.

Problem:
{problem}

Return format:
Strategy 1:
Strategy 2:
Strategy 3:
"""
)

thought_chain = thought_prompt | llm | parser

# -------------------------------
# STEP 2: Evaluate Strategies
# -------------------------------
evaluation_prompt = PromptTemplate(
    input_variables=["problem", "strategies"],
    template="""
You are an expert evaluator.

Evaluate the following strategies for correctness, efficiency, and simplicity.
Select the BEST strategy and justify briefly.

Problem:
{problem}

Strategies:
{strategies}

Best Strategy:
Reason:
"""
)

evaluation_chain = evaluation_prompt | llm | parser

# -------------------------------
# STEP 3: Solve Using Best Strategy
# -------------------------------
solution_prompt = PromptTemplate(
    input_variables=["problem", "best_strategy"],
    template="""
Solve the problem using the selected strategy.

Provide:
1. Clear explanation
2. Final Python code
3. Time & space complexity

Do NOT expose internal reasoning.

Problem:
{problem}

Selected Strategy:
{best_strategy}

Solution:
"""
)

solution_chain = solution_prompt | llm | parser

# -------------------------------
# Main Agent Function
# -------------------------------
def solve_problem(problem: str):
    strategies = thought_chain.invoke({"problem": problem})
    evaluation = evaluation_chain.invoke({
        "problem": problem,
        "strategies": strategies
    })
    final_solution = solution_chain.invoke({
        "problem": problem,
        "best_strategy": evaluation
    })

    return {
        "Generated Strategies": strategies,
        "Evaluation": evaluation,
        "Final Solution": final_solution
    }
```
