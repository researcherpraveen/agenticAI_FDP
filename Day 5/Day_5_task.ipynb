```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image

# -----------------------------
# CONFIGURATION
# -----------------------------
MODEL_ID = "runwayml/stable-diffusion-v1-5"
OUTPUT_IMAGE = "concept_image.png"

TOPIC = "Python Lists and List Comprehension"

PROMPT = f"""
Educational infographic explaining {TOPIC}.
Clean vector style.
Flowchart layout.
Clear sections.
Minimal colors.
White background.
Readable text blocks.
Diagram showing:
- Python Lists
- List Comprehension syntax
- Loop, condition, expression
- Example transformation
Professional academic style.
"""

NEGATIVE_PROMPT = """
photorealistic, human faces, blurry, distorted text,
low quality, noisy, messy layout
"""

# -----------------------------
# LOAD MODEL
# -----------------------------
pipe = StableDiffusionPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16
)

device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = pipe.to(device)

# -----------------------------
# GENERATE IMAGE
# -----------------------------
image = pipe(
    prompt=PROMPT,
    negative_prompt=NEGATIVE_PROMPT,
    height=768,
    width=1024,
    guidance_scale=7.5,
    num_inference_steps=30
).images[0]

# -----------------------------
# SAVE IMAGE
# -----------------------------
image.save(OUTPUT_IMAGE)

print(f"Concept image generated successfully: {OUTPUT_IMAGE}")
```

```python
import os
import torch
from crewai import Agent, Task, Crew
from langchain_groq import ChatGroq
from diffusers import StableDiffusionPipeline
from PIL import Image

# -------------------------------
# CONFIG
# -------------------------------
TOPIC = input("Enter programming topic: ")
IMAGE_NAME = "concept_image.png"

# Set GROQ_API_KEY
os.environ["GROQ_API_KEY"] = "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" # <<< REPLACE WITH YOUR ACTUAL GROQ API KEY
# Removed os.environ["OPENAI_API_KEY"] = "DUMMY_KEY_FOR_CREWAI" as it was causing authentication errors.

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0.2
)

# -------------------------------
# AGENT 1: DEVELOPER AGENT
# -------------------------------
developer_agent = Agent(
    role="Developer Agent",
    goal=f"Solve and explain coding problems related to {TOPIC}",
    backstory="Expert software developer and programming instructor",
    llm=llm,
    verbose=True
)

developer_task = Task(
    description=f"""
Provide a clear explanation and a working code example
for the programming topic: {TOPIC}.
Include comments and best practices.
""",
    expected_output="Code solution with explanation",
    agent=developer_agent
)

# -------------------------------
# AGENT 2: QUIZ AGENT
# -------------------------------
quiz_agent = Agent(
    role="Quiz Agent",
    goal=f"Generate quiz questions on {TOPIC}",
    backstory="Assessment designer and computer science educator",
    llm=llm,
    verbose=True
)

quiz_task = Task(
    description=f"""
Create 10 MCQ questions on {TOPIC}.
Each question must have:
- 4 options (A, B, C, D)
- Correct answer clearly mentioned
""",
    expected_output="10 MCQs with answers",
    agent=quiz_agent
)

# -------------------------------
# AGENT 3: IMAGE GENERATOR AGENT
# -------------------------------
def generate_image(topic):
    pipe = StableDiffusionPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5",
        torch_dtype=torch.float16
    )
    pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")

    prompt = f"Conceptual diagram or flowchart explaining {topic}, clean, educational, minimal design"

    image = pipe(prompt).images[0]
    image.save(IMAGE_NAME)
    return IMAGE_NAME

image_agent = Agent(
    role="Image Generator Agent",
    goal="Generate a conceptual image for learning",
    backstory="Visual AI expert creating educational diagrams",
    llm=llm,
    verbose=True
)

image_task = Task(
    description=f"""
Generate a conceptual image or flowchart
that visually represents the programming topic: {TOPIC}.
""",
    expected_output="Conceptual image generated",
    agent=image_agent
)

# -------------------------------
# CREW EXECUTION
# -------------------------------
crew = Crew(
    agents=[developer_agent, quiz_agent, image_agent],
    tasks=[developer_task, quiz_task, image_task],
    verbose=True
)

result = crew.kickoff()

# -------------------------------
# IMAGE GENERATION (Diffusers)
# -------------------------------
print("\nGenerating conceptual image...")
image_path = generate_image(TOPIC)

# -------------------------------
# OUTPUT
# -------------------------------
print("\n================ FINAL OUTPUT ================\n")
print(result)
print(f"\nConceptual Image Saved As: {image_path}")
```
